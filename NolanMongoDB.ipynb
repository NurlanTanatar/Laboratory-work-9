## Printing out all records in collection

## printing records without ids

```python
for i in mycol.find({},{ "_id": 0, "name": 1, "address": 1 }):
  print(i) 
```

## printing record without address coloumn

```python
for x in mycol.find({},{ "address": 0 }):
  print(x)
```

```python
for i in mycol.find({},{ "_id": 0, "address": 0}):
  print(i)
```

```python
for i in mycol.find({},{ "_id": 0,"name": 1}):
  print(i)
```

## Finding address that starts from S with help of RegEX

```python
myquery = { "address": { "$regex": "^S" } }
mydoc = mycol.find(myquery)
```

```python
for i in mydoc:
  print(i)
```

## sort the result by ascending order

```python
mydoc = mycol.find().sort("name") 
for i in mydoc:
  print(i)
```

## sort the result by descending order

```python
mydoc = mycol.find().sort("name", -1)
for i in mydoc:
  print(i)
```
